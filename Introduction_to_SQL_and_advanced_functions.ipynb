```python
#1 explain the fundamental differences between DDL,DML and DQL commands in SQL.Provide one example for each type of command.
'''DDL commands are used to define the database schema and structure.DDL commands are used to create, modify, and delete database objects such as tables, indexes, views, and stored procedures.
Example: CREATE TABLE customers (id INT PRIMARY KEY, name VARCHAR(50), email VARCHAR(50));
DML commands are used to manipulate the data stored in the database.DML commands are used to insert, update, and delete data from tables.
Example: UPDATE customers SET name = 'John Doe' WHERE id = 1;
DQL commands are used to retrieve data from the database.DQL commands are used to select data from one or more tables.
Example: SELECT * FROM customers WHERE name = 'John Doe';
'''
```

```python
#2 What is the purpose of SQL constraints?Name and describe these common types of constraints,providing a simple scenario where each would be useful.
'''SQL constraints are used to define the rules for the data in a table.
purpose:constraints are used to prevent invalid data from being entered into the database.
three types of constraints:
1.NOT NULL: Ensures that a column cannot have a NULL value.
2.UNIQUE: Ensures that all values in a column are different.
3.PRIMARY KEY: A combination of NOT NULL and UNIQUE. Uniquely identifies each row in a table.
Example scenario: Suppose we have a table of students with columns for their name, age, and grade
and we want to ensure that each student has a unique name. We can add a UNIQUE constraint to the name column like this:
ALTER TABLE students ADD CONSTRAINT unique_name UNIQUE (name);
'''
```

```python
#3 Explain difference between LIMIT and OFFSET clauses in SQL.How would you use them together to retrieve the third page of results,assuming each page has 10 records.
'''LIMIT and OFFSET are used in SQL to limit the number of rows returned by a query.
LIMIT: The LIMIT clause specifies the maximum number of rows to return.
OFFSET: The OFFSET clause specifies the number of rows to skip before returning the results.
Example: Suppose we have a table of customers with columns for their name, age, and email address. We want to retrieve the third page of results, assuming each page has
10 records. We can use the following query to achieve this:
SELECT * FROM customers LIMIT 10 OFFSET'''
```

```python
#4 What is a Common Table Expression (CTE) in SQL,and what are its main benefits?provide a simple SQL example demonstrating its usage.
'''A Common Table Expression (CTE) is a temporary result set that can be used in a SQL query.
It is used to simplify a complex query by breaking it down into smaller, more manageable parts and creating a named
result set that can be referenced in the main query
 Suppose we have a table of employees with columns for their name, age, and salary. We want to retrieve the top 3 highest paid employees
based on their salary
Example:
WITH top_employees AS (
  SELECT name, salary
  FROM employees
  ORDER BY salary DESC
  LIMIT 3
)
SELECT * FROM top_employees;'''
```

```python
#5 Describe the concept of SQL Normalization and its primary goals.Briefly explain the first three normal forms (1NF,2NF,3NF).
'''SQL normalization is the process of organizing data in a database to reduce redundancy and improve data integrity.
Goals of normalization:
1.Eliminate data redundancy
2.Ensure data integrity
3.Improve performance
Normal forms:
1.1NF(first normal form):
-No repeating groups.
-Atomic values in each column.
2.2NF(second normal form):
-Satisfies 1NF.
-No partial dependencies.
3.3NF(third normal form):
-Must be in 2NF
-No transitive dependency.'''
```

```python
#7 Generate a report CustomerName,Email and the TotalNumberOrders for each customer.Include customers who have not placed any orders,in which case their TotalNumbersofOrders should be 0.Order the results by CustomerName.
'''SQL
SELECT customers.name AS CustomerName, customers.email AS Email, COUNT(orders.id) AS TotalNumberOrders
FROM customers
LEFT JOIN orders ON customers.id = orders.customer_id
GROUP BY customers.id
ORDER BY CustomerName;'''
```

```python
#8 Retreive Product Information with category:Write a SQL query to display the ProductName,Price,StockQuantity,and CategoryName dor all Products.Order the results by CategoryName and then ProductName alphabetically.
'''SQL
SELECT products.name AS ProductName, products.price AS Price, products.stockquantity AS StockQuantity, categories.name AS CategoryName
FROM products
JOIN categories ON products.category_id = categories.id
ORDER BY CategoryName, ProductName;'''
```

```python
#9 Write a SQL query that uses a Common Table Expression (CTE) and a Window Function(specifically ROW_NUMBER()or RANK())to display the CategoryName,ProductsName and Price for the top 2 most expensive products in each CategoryName.
'''SQL
WITH top_products AS (
  SELECT categories.name AS CategoryName, products.name AS ProductName, products.price AS Price,
         ROW_NUMBER() OVER (PARTITION BY categories.name ORDER BY products.price DESC) AS price_rank
  FROM categories
  JOIN products ON categories.id = products.category_id
)
Select CategoryName, ProductName, Price
FROM top_products
WHERE price_rank <= 2;'''
```
